### Total - step1

```python
## Total - step1
import pandas as pd
#pd.set_option('display.max_rows', None)

datafile = 'https://raw.githubusercontent.com/Soyoung-Yoon/test/main/data_01.xlsx'
df = pd.read_excel(datafile, skiprows=2)

def get_target_data(df):
    cond0 = df['출장비지급']=='지급'
    cond1 = (df['상태'] == '완료') & (df['취소상태'] == '완료')
    cond2 = df['상태'] == '등록'
    cond3 = df['상태'] == '회수'
    df2 = df[cond0 & ~(cond1 | cond2 | cond3)].copy()
    return df2

df2 = get_target_data(df)
#df2['출장시간'] = df2['일수'].str.split('시').str[0].astype('int32')
df2['취소상태'] = df2['취소상태'].fillna('')
df2['출장시간'] = df2['일수'].str[0:2].astype('int32')
df2['대중교통금액'] = df2['출장시간'].map(lambda x : 10000 if x < 4 else 20000)
df2['비고'] = df2['이동수단'].map(lambda x : '행정 차량의 경우 -1만원' if x == '행정차량' else '')
df2['대중교통금액'] = df2['대중교통금액'] - df2['이동수단'].map(lambda x : 10000 if x == '행정차량' else 0)
df2['월'] = df2['시작일'].str.split('.').str[1]
df2['일'] = df2['시작일'].str.split('.').str[2].str.split('[').str[0]
temp = df2['출장시간'].copy()
df2 = df2.drop(columns=['출장시간'])
df2['출장시간'] = temp
df2['차감액'] = 0
df2['실지급액'] = 0
df2 = df2.sort_values(['대상자', '월', '일'], ignore_index=True)
df2.loc[(df2['비고'] !=''), ['대상자', '시작일', '일수', '이동수단', '출장시간', '대중교통금액', '비고']]
```

### Total - step2

```python
## Total - step2
temp = df2.groupby(['대상자', '월', '일'])
limit_day = 20000
processed = []
for name, gtemp in temp:
    if len(gtemp) > 1:
        total = gtemp["대중교통금액"].sum()
        if total> limit_day:
            deductible_amount = total - limit_day
            gtemp.iloc[-1, -6] = f'1일 한도(2만 원) 초과로 {deductible_amount}원 미인정'
            gtemp.iloc[-1, -2] = deductible_amount
            display(name, gtemp)
    processed.append(gtemp)

# 1일 한도 초과 처리 완료된 데이터 (비고에 표시)
df_day = pd.concat(processed)
print(df_day.shape)
```

### Total - step3

```python
### Total - step3
limit_month = 150000
df_month = df_day.groupby(['대상자', '월'])[['대중교통금액', '차감액']].sum(numeric_only=True).reset_index()
df_month['실지급액'] = df_month['대중교통금액'] - df_month['차감액']
df_month['비고'] = ''
processed = []
for idx in range(len(df_month)):
    temp = df_month.loc[[idx], :].copy()
    if temp.loc[idx, '실지급액'] > limit_month:
        temp.loc[idx, '실지급액'] = limit_month
        orginal_amount = temp.loc[idx, '대중교통금액']
        over_amount = orginal_amount - limit_month
        temp.loc[idx, '차감액'] += over_amount
        temp['비고'] = f'월 한도 초과({orginal_amount:,})로 {over_amount//10000}만원 차감'
    processed.append(temp)
df_month = pd.concat(processed)
df_person = df_month.groupby(['대상자'])[['실지급액']].sum(numeric_only=True).reset_index()
display(df_month, df_person)
```

```python
## Total - step4
colnames = ['월', '상태', '취소상태', '출장종류', '등록자', '대상자', '시작일', '시작시간', '종료일', '종료시간',
            '일수', '출장지', '출장지주소', '출장목적', '이동수단', '출장비지급', '대중교통금액', '비고']
df_day2 = df_day[colnames]
#display(df_day2.head(10))  # original

def make_month(person, idx_m_start):
    person_month = []
    month = person.groupby(['대상자', '월'])
    for idx_m, (name, mtemp) in enumerate(month, start=idx_m_start):
        msg = f'{int(df_month.loc[idx_m, "월"])}월 합계'
        dummy = [[msg] + ['']*15 + [df_month.loc[idx_m, '실지급액']] + [df_month.loc[idx_m, '비고']]]
        total_month = pd.DataFrame(dummy, columns=colnames, index=[9999])
        #display(total_month)
        #display(pd.concat([gtemp, total_df]))
        person_month.append(pd.concat([mtemp, total_month]))
    return idx_m + 1, pd.concat(person_month)


processed_p = []
idx_m_start = 0
person_groupby = df_day2.groupby(['대상자'])
for idx_p, (pname, person) in enumerate(person_groupby):
    idx_m_start, df_temp = make_month(person, idx_m_start)
    dummy = [['총 지급액'] + ['']*15 + [df_person.loc[idx_p, '실지급액']] + ['']]
    total_person = pd.DataFrame(dummy, columns=colnames, index=[99999])
    processed_p.append(pd.concat([df_temp, total_person]))

df_day3 = pd.concat(processed_p)
df_day3.head(20)
```

```python
writer = pd.ExcelWriter('Lesson01_Result_240419.xlsx')
df_day3.to_excel(writer, sheet_name='지급목록', index=False)
df_month.to_excel(writer, sheet_name='월별', index=False)
df_person.to_excel(writer, sheet_name='대상자별', index=False)
writer.close()
```
